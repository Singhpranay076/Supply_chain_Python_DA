```python
import pandas as pd      #pandas to analyze datasets
import numpy as np       #numpy for numerical calculations
```

```python
df = pd.read_csv("supply_chain_data.csv")     #read_csv() is a function in pandas that reads a CSV file.

print(df.head())    #This prints the first rows so you understand the structure.
```

```python
print(df.isnull().sum())          #df is your DataFrame created earlier when you loaded the CSV using pandas.
                                  #isnull() checks every cell in the DataFrame.
                                  #So when .sum() is applied column-wise, it counts how many True values exist.
                            
```

```python
df = df.fillna(0)              #df.dropna()   ---- Remove rows with missing values
                               #df.fillna(0)  ---- Replace missing with 0
```

```python
df["Product type"] = df["Product type"].str.lower()           

#df acts as the main object through which you interact with the dataset.
#produces a new Series where every value is lowercase.
```

```python
total_revenue = df["Revenue generated"].sum()    #Calculates the total revenue from the dataset and stores the result in a variable

print("Total Revenue:", total_revenue)
```

```python
for index, row in df.iterrows():           #iterrows() means iterate through rows of the DataFrame.
                                            #for creates a loop.

    stock = row["Stock levels"]
    product = row["SKU"]

    if stock < 20:
        print(product, "Low stock")
```

```python
demand_std = np.std(df["Number of products sold"])          #np.std() calculates the standard deviation of numbers using numpy

print("Demand variability:", demand_std)    
```

```python
def check_reorder(stock, demand):           

    reorder_point = demand * 0.5

    if stock < reorder_point:
        return "Reorder Needed"
    else:
        return "Stock OK"
```

```python
df["Reorder Status"] = df.apply(lambda x: check_reorder(x["Stock levels"], x["Number of products sold"]),
    axis=1
)

print(df[["SKU","Reorder Status"]])

#Note: 1. lambda creates a small anonymous function (a function without a name).
#      2. Creates a new column that tells whether a product needs to be reordered based on stock and sales. 
```

```python
class Product:

    def __init__(self, sku, stock, demand):
        self.sku = sku
        self.stock = stock
        self.demand = demand

    def reorder_check(self):

        if self.stock < self.demand * 0.5:
            print(self.sku, "Reorder required")
        else:
            print(self.sku, "Stock sufficient")
```

```python
products = []

for index, row in df.iterrows():

    p = Product(row["SKU"],
                row["Stock levels"],
                row["Number of products sold"])

    products.append(p)

for p in products:
    p.reorder_check()
```
